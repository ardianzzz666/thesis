```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('data/yogyakarta/weather_yogya.csv')
df = weather
weather.head()
```

```python
del df['avt']
del df['mxt']
del df['mnt']

df.head()
```

```python
df.date = pd.to_datetime(df.date)
df.set_index('date', inplace=True)

df.head()
```

```python
df.plot()
plt.show()
```

```python
new_df = df.interpolate(method='time', order=2)
new_df.plot()
plt.show()
```
